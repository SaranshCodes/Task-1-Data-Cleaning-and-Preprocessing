```python
df["Date Added"]=pd.to_datetime(df["Date Added"],errors="coerce")
```

```python
df["Imdb Score"]=df["Imdb Score"].str.replace("/10","")
df["Imdb Score"]=pd.to_numeric(df["Imdb Score"])
```

```python
df.rename(columns={"Release Date":"Release Year"},inplace=True)
```

```python
df.isnull().sum()
```

```python
missing_values=df.isnull().mean()*100
print(missing_values)
```

```python
string_fields=df.select_dtypes(include="object").columns
df[string_fields]=df[string_fields].fillna("Not Available")
```

```python
duplicates=df.duplicated().sum()
print(duplicates)
```

```python
df["Title"]=df["Title"].str.replace("#","")
```

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
```

```python
df["release_year"]=df["release_year"].astype("Int64")
```
